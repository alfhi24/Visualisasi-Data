```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

data : https://raw.githubusercontent.com/alfhi24/Visualisasi-Data/main/Network%20Visualization/jabar.txt

```python
data = 'jabar.txt'
dataGraph = nx.read_edgelist(data)
type (dataGraph)
print(nx.info(dataGraph))
```

```python
# List edge yang terbentuk
edgeList = pd.read_csv(data, sep="\t", header=None)
edgeList.head()
```

```python
# Visualisasi graph
G = nx.Graph()
for i, edge in edgeList.iterrows():
    G.add_edge(edge[0], edge[1])
nx.draw(G, with_labels = True)
plt.show()
```

```python
# Menghitung degree centrality
nx.degree_centrality(G)
```

```python
list(G.neighbors("'Sumedang'"))
```

Berdasarkan hasil di atas, maka kota yang memiliki degree centrality paling besar adalah Sumedang, artinya kota Sumedang memiliki paling banyak kota tetangga

```python
# Menghitung betweenes centrality
nx.betweenness_centrality(G)
```

Berdasarkan hasil di atas, maka kota yang memiliki betweeness centrality paling besar adalah Bandung, artinya kota Bandung adalah kota yang paling sering dilewati untuk berpergian dari suatu kota ke kota lain

```python
list(nx.find_cliques(G))
```
